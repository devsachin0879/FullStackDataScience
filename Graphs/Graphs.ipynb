```python
from mpl_toolkits import mplot3d
```

```python
x = np.linspace(-1,6,30)
y = np.linspace(-1,6,30)
```

```python
def sin_func(x,y):
    return np.sin(np.sqrt(x**2+y**2))
```

```python
z = sin_func(x,y)
```

```python
ax = plt.axes(projection = '3d')
ax.plot3D(x,y,z)
```

```python
x = np.linspace(-1,6,30)
y = np.linspace(-1,6,30)
z = x+y
```

```python
ax = plt.axes(projection = '3d')
ax.plot3D(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
