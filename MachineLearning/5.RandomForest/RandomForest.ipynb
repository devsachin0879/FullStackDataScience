```python
import pandas as pd
import numpy as np
df=pd.read_csv("https://raw.githubusercontent.com/shrikant-temburwar/Wine-Quality-Dataset/master/winequality-red.csv",sep = ';')
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
X = df.drop('quality',axis=1)
```

```python
y = df['quality']
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(25,15))
tree.plot_tree(model,filled=True)
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_predict)
```

```python
grid_param = {
    'criterion': ['gini', 'entropy'],
    'max_depth' : range(2,10,1),
    'min_samples_leaf' : range(1,8,1),
    'min_samples_split': range(2,8,1),
    'splitter' : ['best', 'random']
    
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=grid_param, cv=3, verbose=1)
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
model_with_best_params=DecisionTreeClassifier(criterion= 'gini',max_depth= 9,min_samples_leaf= 7,min_samples_split= 4,splitter='random')
```

```python
model_with_best_params.fit(X_train,y_train)
```

```python
y_predict2 = model_with_best_params.predict(X_test)
```

```python
accuracy_score(y_test,y_predict2)
```

##### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
Rf_model = RandomForestClassifier()
```

```python
Rf_model.fit(X_train,y_train)
```

```python
y_pred_rf = Rf_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred_rf)
```

```python
# we are tuning three hyperparameters right now, we are passing the different values for both parameters
grid_param = {
    "n_estimators" : [90,100,115,130],
    'criterion': ['gini', 'entropy'],
    'max_depth' : range(2,20,1),
    'min_samples_leaf' : range(1,10,1),
    'min_samples_split': range(2,10,1),
    'max_features' : ['auto','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(estimator=Rf_model,param_grid=grid_param,cv=3,verbose=2,n_jobs=-1)
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
Rf_model_with_best_params=RandomForestClassifier(criterion='entropy',max_depth= 13,max_features= 'auto',min_samples_leaf= 1,min_samples_split= 2,n_estimators=115)
```

```python
Rf_model_with_best_params.fit(X_train,y_train)
```

```python
y_predict_rf_bp = Rf_model_with_best_params.predict(X_test)
```

```python
accuracy_score(y_test,y_predict_rf_bp)
```
