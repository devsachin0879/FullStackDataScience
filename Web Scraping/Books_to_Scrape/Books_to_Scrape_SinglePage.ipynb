```python
url1 = "http://books.toscrape.com/catalogue/page-1.html"
url1
```

```python
## check of we are able to hit the url using python
response_website = urReq(url1)
response_website
```

```python
data_book = response_website.read()
data_book
```

```python
### parse data in html format
beautified_html = bs(data_book,'html.parser')
beautified_html
```

```python
## in the page we have list of items
big_box = beautified_html.find_all("li",{"class":"col-xs-6 col-sm-4 col-md-3 col-lg-3"})
big_box
```
