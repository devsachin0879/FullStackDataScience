```python
from flask import Flask, render_template, jsonify, request
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
import requests
import pandas as pd
import csv
```

```python
try:
    for i in range(1,51):
        url = "http://books.toscrape.com/catalogue/page-" + str(i)+".html"
        uClient = uReq(url)
        mainPage = uClient.read()
        uClient.close()

        mainPage_html = bs(mainPage,"html.parser")

        big_box = mainPage_html.find_all("li",{"class":"col-xs-6 col-sm-4 col-md-3 col-lg-3"})
        filename = "BookstoScraper.csv"
        #fw = open(filename,"w")
        #headers = "Title, Price, Category \n"
        #fw.write(headers)
        data_list = []

        for prod_num in range(len(big_box)):
            book_link = "http://books.toscrape.com/catalogue/" + big_box[prod_num].a['href']
            bookRes = requests.get(book_link)

            ### encoding
            bookRes.encoding = 'utf-8'
            book_html = bs(bookRes.text,"html.parser")
            # print(book_html)

            title = book_html.find_all('h1')[0].text
            price = book_html.find_all("p",{"class":"price_color"})[0].text
            category = book_html.find_all("a")[3].text

            mydict = {"Title":title, "Price in Pounds":price[1:], "Category":category}
            data_list.append(mydict)
        df = pd.DataFrame(data_list)
        if i == 1:
            df.to_csv(filename,mode='a', index=False, header=True)
        else:
            df.to_csv(filename,mode='a', index=False, header=False)
except Exception as e:
    print("An error has occurred.",e)
```
