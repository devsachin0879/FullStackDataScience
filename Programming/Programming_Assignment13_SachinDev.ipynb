```python
import math
def compute_value():
    C = 50
    H = 30
    D = eval(input("PLease enter values for D: "))
    res = []
    for i in D:
        sol = int(math.sqrt((2*C*i/H))) ##using int to remove decimals
        res.append(str(sol))
    print('Output: {}'.format(','.join(res)))
```

```python
compute_value()
```

```python
def generateArray():
    x = int(input("Please enter num of sub arrays: "))
    y = int(input("Please enter num of cols in each array: "))
    l = []
    for i in range(x):
        l.insert(i,[])
        for j in range(y):
            l[i].append(i*j)
    print(l)
```

```python
def sort_string():
    s = input("Please enter a sentence: ")
    s = s.split(",")
    out = ','.join(sorted(s))
    print(f"Output: {out}")
```

```python
sort_string()
```

```python
def sort_alphba():
    s = input("Please enter the string: ").split(" ")
    print("\n")
    ##remove duplicates
    s =set(s)
    s = sorted(s)
    out = ' '.join(s)
    print(f"output:\n{out}")
```

```python
import re
def calc_char_num():
    s = input("PLease enter  the string: ")
    print("\n")
    
    ### find only letters
    pattern = re.compile(r"[A-Za-z]")
    matches = pattern.finditer(s)
    
    ## append letters in empty list and then find length
    l = []
    for i in matches:
        l.append(i)
    length_string = len(l)
    
    ## append numbers in empty list and then find length
    pattern1 = re.compile(r"[0-9]")
    matches1 = pattern1.finditer(s)                      
    l1 = []
    for j in matches1:
        l1.append(j)
    length_nums = len(l1)     
                          
    print(f"Letter: {length_string}")
    print(f"Numbers: {length_nums}")
```

```python
calc_char_num()
```

```python
def check_password():
    pswd = input("Please enter password: ").split(",")
    
    l = []
    for i in pswd:
        if re.fullmatch(r"^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$#])[A-Za-z\d@$#]{6,12}$",i):
            l.append(i)
    for j in l:
        print(j)
```

```python
check_password()
```
