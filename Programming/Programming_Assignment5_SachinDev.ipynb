###### 1.	Write a Python Program to Find LCM?

```python
def find_lcm(x,y):
    ###check the higher number
    if x > y:
        higher = x
    else:
        higher = y
    
    ### assigning higher value to a variable
    var = higher
    
    while True:
        if (higher%x == 0 and higher%y == 0):
            print(higher)
            break
        else:
            higher = higher + var
    
```

###### 2.	Write a Python Program to Find HCF?

```python
def find_hcf(a,b):
    if b == 0:
        return a
    else:
        return find_hcf(b,a%b)
```

```python
find_hcf(12,64)
```

```python
find_hcf(12,18)
```

###### 3.	Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

```python
def convert_dec(a):
    print(f"Binary Value: {bin(a)}")
    print(f"Octal Value: {oct(a)}")
    print(f"Hexadecimal Value: {hex(a)}")
```

```python
convert_dec(344)
```

```python
convert_dec(122)
```

###### 5.	Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

```python
def calc():
    a = int(input("Please enter first number: "))
    b = int(input("Please enter second number: "))
    oper = input("Enter the operator +,-,* or /: ")
    
    if oper == '+':
        return a+b
    elif oper == '-':
        return a-b
    elif oper == '*':
        return a*b
    elif oper == '/':
        return a/b
    else:
        return"Please check the operator or numbers"
```

```python
calc()
```
