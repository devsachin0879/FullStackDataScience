```python
# check description for LTV column
df["LTV rate"].describe()
```

On an average LTV rate is 75.4%

Most of the finance companies keep 80% LTV ratio as threshold, anything below that is even better. If LTV rate rises above 80% borrowing costs can become higher, or borrowers may be denied loans 

```python
## flag ltv rates
def ltv_flag(a):
    if a > 0.8:
        return 'Bad'
    else:
        return 'Good'
```

```python
##creating a new column LTV Flag
df['LTV Flag'] = df['LTV rate'].apply(ltv_flag)
df.head()
```

```python
##check values of ltv flag column
df['LTV Flag'].value_counts()
```

More than 39% customers paid less than 20% of the loan amount as down payment.

```python
### check cusotmer where ltv flag is bad
df[['Loan amount (issued amount)','Interest','Total payable amount','Monthly payment','Incoming repayments total','Expected repayments total']][df['LTV Flag'] == 'Bad'].describe()
```

Here we are able to get only 34.75% of the expected repayments.
