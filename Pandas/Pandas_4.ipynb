```python
##if profit is +ve then profitflag = POSITIVE else profit_flag = NEGATIVE
```

```python
def profit_flag(a):
    if a>=0:
        return 'Positive'
    else:
        return 'Negative'
```

```python
profit_flag(-26)
```

```python
profit_flag(26)
```

```python
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
##create column to store length of customer name
df_sales['len_cyst_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
df_sales['quantity'].unique()
```

```python
## qunatity b/w 0 to 5 low, 6-10 meduim , above 10 high
def quantity_flag(a):
    if a < 5:
        return "low"
    elif a>5 and a<11:
        return "medium"
    else:
        return "high"
```

```python
df_sales["quantity_flag"] = df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head()
```

```python
##using lambda func
df_sales["quantity_sqaure"] = df_sales['quantity'].apply(lambda x: x**2)
```
