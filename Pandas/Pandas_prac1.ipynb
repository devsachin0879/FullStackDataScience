```python
### remove $ from item_price

df['item_price'] = df['item_price'].str.replace("$","")
```

```python
df.head()
```

```python
df['item_price'] = df['item_price'].str.replace(" ","")
```

```python
### before multiplying quantity and item_price for revenue, we need to convert item_price from str to int
df = df.astype({'item_price':'float'})
```

```python
df.dtypes
```

```python
df['revenue'] = df['quantity'] * df['item_price']
```

```python
df.head()
```

```python
### item with higest revenue

df[df['revenue'] == max(df['revenue'])]
```

```python
### item with lowest revenue

ser = df['revenue'][df['revenue'] == min(df['revenue'])]
```

```python
item_count_low_rev = len(df.iloc[ser.index])
```

```python
item_count_low_rev
```
