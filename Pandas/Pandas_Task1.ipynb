```python
import pandas as pd
import csv
import mysql.connector as connection
import os
```

```python
conn = connection.connect(host='localhost',user='root',passwd='MSdhoni07',use_pure=True)
cur = conn.cursor()

##read data from csv
csv_data = csv.reader(open('FitBit_data.csv'))
cur.execute('CREATE TABLE sachin12345.FitBit_task (Id FLOAT NOT NULL,ActivityDate DATE NOT NULL,TotalSteps FLOAT NOT NULL,\
TotalDistance FLOAT NOT NULL,TrackerDistance FLOAT NOT NULL,LoggedActivitiesDistance FLOAT NOT NULL,\
VeryActiveDistance FLOAT NOT NULL,ModeratelyActiveDistance FLOAT NOT NULL,LightActiveDistance FLOAT NOT NULL,\
SedentaryActiveDistance FLOAT NOT NULL,VeryActiveMinutes FLOAT NOT NULL,FairlyActiveMinutes FLOAT NOT NULL,\
LightlyActiveMinutes FLOAT NOT NULL,SedentaryMinutes FLOAT NOT NULL,Calories FLOAT NOT NULL)')

header = next(csv_data)
for row in csv_data:
    print(row)
    cur.execute("Insert into sachin12345.FitBit_task(Id,ActivityDate,TotalSteps,TotalDistance,TrackerDistance,\
    LoggedActivitiesDistance,VeryActiveDistance,ModeratelyActiveDistance,LightActiveDistance,SedentaryActiveDistance,\
    VeryActiveMinutes,FairlyActiveMinutes,LightlyActiveMinutes,SedentaryMinutes,Calories) \
    VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)",row)
conn.commit()
cur.close()
print("All Data Uploaded")
```

###### 8. Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset

```python
## on an average 2000 to 2500 calories should be burned
df_cal = df.groupby('Id')['Calories'].mean()
df_cal[df_cal >= 2000].count()
```

###### 9. how many person are not a regular person with respect to activity try to find out those

```python
### A perosn is regular if he is active for atleast 14 days
df_active = df.groupby('Id')['ActivityDate'].count()
df_active[df_active >= 14].count()
```

###### 10. who is the thired most active person in this dataset find out those in pandas and in sql both 

```python
df_third_active = pd.DataFrame(df.groupby('Id')['ActivityDate'].count().sort_values(ascending=False))
df_third_active.iloc[2]
```

###### 11. who is the 5th most laziest person avilable in dataset find it out

```python
df_lazy = pd.DataFrame(df.groupby('Id')['TotalDistance'].sum().sort_values(ascending=True))
df_lazy.iloc[4]
```

###### 12. what is a totla acumulative calories burn for a person find out

```python
df.groupby('Id')['Calories'].cumsum()
```
