###### 9. Average Weekly response time for each agent

```python
df_weekly.head()
```

```python
agent_perf['Date_Week'] = pd.to_datetime(agent_perf['Date']).dt.week
```

```python
agent_perf.head()
```

```python
def time_convert(x):
    times = x.split(':')
    return(60* int(times[0]) + 60*int(times[1])) + int(times[2])
```

```python
agent_perf['Avg_res_time_dur'] = agent_perf['Average Response Time'].apply(time_convert)
```

```python
agent_perf.head()
```

```python
agent_perf.groupby(['Date_Week','Agent Name'])['Avg_res_time_dur'].mean()
```

###### 12. % of chat on which they have received feedback

```python
agent_perf.head()
```

```python
df2 = agent_perf.groupby('Agent Name')['Total Chats'].sum().reset_index()
df2.head()
```

```python
df2['Total_chats'] = agent_perf['Total Chats'].sum()
```

```python
df2.head()
```

```python
df2 = df2[df2['Total Chats'] !=0]
```

```python
df2['%Chats'] = (df2['Total Chats']/df2['Total_chats'])*100
```

```python
df2.head()
```

###### 13: Total contribution hour for each and every agents weekly basis

```python
agent_log.head()
```

```python
agent_log['Date_week'] = pd.to_datetime(agent_log['Date']).dt.week
```

```python
agent_log['Duration_hrs'] = agent_log['Duration'].str.split(':',expand=True)[0]
agent_log['Duration_mins'] = agent_log['Duration'].str.split(':',expand=True)[1]
agent_log['Duration_sec'] = agent_log['Duration'].str.split(':',expand=True)[2]

agent_log['Duration_hrs'] = agent_log['Duration_hrs'].astype(int) + agent_log['Duration_mins'].astype(int)/60 + agent_log['Duration_sec'].astype(int)/(60*60)
agent_log.groupby(['Date_week','Agent'])['Duration_hrs'].mean()
```

###### 14. Total %of active hour for a month

```python
agent_log['Date_Month'] = pd.to_datetime(agent_log['Date']).dt.month
agent_log.head()
```

```python
df2 = agent_log.groupby(['Date_Month','Agent'])['Duration_hrs'].sum().reset_index()
```

```python
df2.head()
```

```python
## total working hrs in a shift = 8hrs
## total days worked by each agent = 20
df2['Total_shift_hrs'] = 8*20
```

```python
df2.head()
```

```python
df2['%active_hrs'] = (df2['Duration_hrs']/df2['Total_shift_hrs'])*100
```
