##### Check Normality of data

If p-value is less than 0.05 then feature is normally distributed.

```python
from scipy.stats import normaltest
```

```python
normaltest(data['target'])
```

```python
normaltest(data['target'])[1]
```

```python
## check p-value for each numeric feature

for feature in num_col:
    normality = normaltest(data[feature])[1]*100
    print(f"P-Value for {feature} is {normality}")
    print("----------------------------------------")
```

##### Check Outliers

IQR = Q3-Q1<br>
Upper Limit = Q3 + 1.5 x IQR<br>
Lower Limit = Q1 - 1.5 x IQR

```python
#### boxplot
for feature in num_col:
    print(feature)
    plt.figure()
    sns.boxplot(data = data[feature])
    plt.show()
```

Seems like there are some outliers in city_index_devlopment and training_hours

```python
#### FOr city_index_development
```

```python
#### get IQR, upper_limit and lower_limit
q1 = data['city_development_index'].quantile(0.25)
q3 = data['city_development_index'].quantile(0.75)
IQR = q3-q1
upper_limit = q3 + (1.5*IQR)
lower_limit = q1 - (1.5*IQR)
```

```python
upper_limit
```

```python
lower_limit
```

```python
### check data less than lower limit and greater than upper limit
data[data['city_development_index']<lower_limit]
```

```python
data[data['city_development_index']<lower_limit].count() ### outliers
```

```python
data[data['city_development_index']>upper_limit]
```

There are 17 outliers in city_development_index column

```python
### check for training_hours

#### get IQR, upper_limit and lower_limit
q1 = data['training_hours'].quantile(0.25)
q3 = data['training_hours'].quantile(0.75)
IQR = q3-q1
upper_limit = q3 + (1.5*IQR)
lower_limit = q1 - (1.5*IQR)
```

```python
upper_limit
```

```python
lower_limit
```

```python
### outliers
data[data['training_hours']>upper_limit]
```

```python
data['training_hours'][data['training_hours']>upper_limit].count()
```

##### Countplot

```python
for feature in ['gender','relevent_experience','enrolled_university','education_level','major_discipline','experience','company_size','company_type','last_new_job','target']:
    print(feature)
    plt.figure()
    sns.countplot(data,x=feature)
    plt.show()
    print("----------------------")
```
