```python
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mpl_dates
```

##### Basic date timeplot

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.figure(figsize=(10,5))
plt.plot_date(dates,y)

plt.tight_layout()
plt.show()
```

##### Change line style

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.figure(figsize=(8,5))
plt.plot_date(dates,y,linestyle='solid')

plt.tight_layout()
plt.show()
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates,y,linestyle='solid')

##gcf = get current figure
## autofmt = auto format
plt.gcf().autofmt_xdate()

plt.tight_layout()
plt.show()
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates,y,linestyle='solid')
plt.gcf().autofmt_xdate()

### format date
date_format = mpl_dates.DateFormatter('%b,%d %Y')
##gca = get current axis
plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.show()
```

##### Real World Data

```python
data = pd.read_csv('data_timeSeries.csv')
data.head()
```

```python
price_date = data['Date']
price_close = data['Close']

plt.figure(figsize=(8,5))
plt.plot_date(price_date,price_close, linestyle='solid')
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()
plt.show()
```
