```python
import matplotlib.pyplot as plt
%matplotlib inline
```

##### Create a Simple Pie Chart

```python
slices = [60,40]

plt.style.use("fivethirtyeight")

###create a pie chart
plt.pie(slices)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

##### Add labels

```python
slices = [60,40]
labels = ['Sixty', 'Forty']

plt.style.use("fivethirtyeight")

###create a pie chart
plt.pie(slices, labels=labels)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

##### Wedge

```python
slices = [60,40]
labels = ['Sixty', 'Forty']

plt.style.use("fivethirtyeight")

### add wedge property
plt.pie(slices, labels=labels,wedgeprops={'edgecolor':'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

##### Slices

##### EXPLODE

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0] ### 0.1 for python as I want to emphasis on python

plt.figure(figsize=(5,5))
plt.style.use("fivethirtyeight")

## explode
plt.pie(slices, labels=labels,explode=explode, wedgeprops={'edgecolor':'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

##### Other Arguments

**Shadow**

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0] ### 0.1 for python as I want to emphasis on python

plt.figure(figsize=(5,5))
plt.style.use("fivethirtyeight")

## add shadow
plt.pie(slices, labels=labels,explode=explode,shadow=True, wedgeprops={'edgecolor':'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

**Starting Angle**

```python
plt.figure(figsize=(5,5))
plt.style.use("fivethirtyeight")

## add angle
plt.pie(slices, labels=labels,explode=explode,shadow=True,startangle=90, wedgeprops={'edgecolor':'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

**Display Percentage**

```python
plt.figure(figsize=(5,5))
plt.style.use("fivethirtyeight")

## add autopct
plt.pie(slices, labels=labels,explode=explode,shadow=True,startangle=90, autopct='%1.1f%%', wedgeprops={'edgecolor':'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.style.use("fivethirtyeight")

## add autopct
plt.pie(slices, labels=labels,explode=explode,shadow=True,startangle=90, autopct='%1.2f%%', wedgeprops={'edgecolor':'black'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
